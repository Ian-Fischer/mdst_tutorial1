```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# feel free to import other libraries! 
```

```python
df = pd.read_csv('../data/states_edu.csv')
df.head()
```

Chosen test: **<hit `Enter` to edit>**

<h2> Cleanup (optional)</h2>

_Use this space to rename columns, deal with missing data, etc._

```python
df['AVG_MATH_8_SCORE'] = df['AVG_MATH_8_SCORE'].replace(np.nan, 0)
df['AVG_MATH_4_SCORE'] = df['AVG_MATH_4_SCORE'].replace(np.nan, 0)
df['AVG_READING_4_SCORE'] = df['AVG_READING_4_SCORE'].replace(np.nan, 0)
df['AVG_READING_8_SCORE'] = df['AVG_READING_8_SCORE'].replace(np.nan, 0)
df['INSTRUCTION_EXPENDITURE'] = df['INSTRUCTION_EXPENDITURE'].replace(np.nan, 0)
df['TOTAL_EXPENDITURE'] = df['TOTAL_EXPENDITURE'].replace(np.nan, 0)

df.head()
```

<h2> Feature Selection </h2>

_Use this space to modify or create features_

```python
address = df['INSTRUCTION_EXPENDITURE']/df['TOTAL_EXPENDITURE']
  
# Using 'Address' as the column name 
# and equating it to the list 
df['Hold'] = address 

hold2 = (df['AVG_MATH_4_SCORE'] + df['AVG_MATH_8_SCORE'] + df['AVG_READING_4_SCORE'] + df['AVG_READING_8_SCORE'])/4

df['Hold2'] = hold2 

df = df[df != 0]

df.head()
```

Final feature list: I think that the most meaningful relationship will be between the percent of revenue to education vs. all grade average test score

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df["Hold"]
y = df["Hold2"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=100, random_state=0)
```

<h2> Prediction </h2>

Chosen ML task: Linear Regression

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression().fit(X,y)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR CLASSIFICATION ONLY:
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model, X_test, y_test,
                         cmap=plt.cm.Blues)
```
